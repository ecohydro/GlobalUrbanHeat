# Make this a function

```python
def df_interp(df, start_col, end_col, start, end):
    """ Runs a linear interpolution between two df columns
    Args:
        df = dataframe
        start = start column
        end = end column
        start = start of num range
        end = end of num range
    """
    
    new_df = pd.DataFrame() # make an empty df
    nan_list = np.full((1, len(df_pop)), np.nan)[0] # make an empty df list

    new_df[start] = df[start_col] # snag the first column
    
    # make a bunch of nan cols
    for i in range(end - start - 1):
        i = i + 1 # fix zero indexing
        column = '19'+str(start+i) #<<<<<---- add 0 if needed
        new_df[column] = nan_list
    
    new_df[end] = df_pop[end_col]
    
    # interp ! ! !
    new_df = new_df.interpolate('linear', axis = 1)
    
    return new_df
```

```python
df7590 = df_interp(df_pop, 'P75', 'P90', 75, 90)

df7590.rename(columns = {75:1975, 90:1990}, inplace = True)
```

```python
df9000 = df_interp(df_pop, 'P90', 'P00', 90, 100)
df9000.rename(columns = {90:1990, 100:2000}, inplace = True)
```

```python
df0015 = df_interp(df_pop, 'P00', 'P15', 0, 15)
df0015.rename(columns = {100:2000, 1}, inplace = True)
```

```python
# join

df_join = df7590.join(df9000.iloc[:,1:])
df_join = df_join.join(df0015.iloc[:,1:])
```
