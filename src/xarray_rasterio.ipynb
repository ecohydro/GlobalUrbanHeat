```python
import xarray 
import multiprocessing as mp 
from multiprocessing import Pool
import os 
import glob
```

```python
def open_rsts(rsts):
    rst_fn1 = rsts[0] 
    rst_fn2 = rsts[1]
    
    "Open two rasters with xarray"
    rst_1 = xarray.open_rasterio(rst_fn1)
    rst_2 = xarray.open_rasterio(rst_fn2)
    print('Done')
```

```python
rh_path = os.path.join('/home/CHIRTS/daily_ERA5/w-ERA5_Td.eq2-2-spechum-Tmax/1983') # RH min made with CHIRTS-daily Tmax
tmax_path = os.path.join('/home/chc-data-out/products/CHIRTSdaily/v1.0/global_tifs_p05/Tmax/1983') # CHIRTS-daily Tmax

rst_fn1_list = glob.glob(rh_path+'/*.tif')
rst_fn2_list = glob.glob(tmax_path+'/*.tif')
zipped_list = list(zip(rst_fn1_list,rst_fn2_list))
```

```python
cpu_num = 4
pool = Pool(processes = cpu_num)
pool.map(open_rsts, zipped_list)
pool.close()
```
